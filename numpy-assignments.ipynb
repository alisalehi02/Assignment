### سوال ۳: مشتق تابع چند متغیره
تابع $f(x,y) = \sin(x \times y)$ را در بازه‌ی $-\pi/2 < x < \pi/2$ و $-\pi/2 < y < \pi/2$ تشکیل دهید.  
با استفاده از `np.gradient`:
1. **مشتق این تابع نسبت به $x$** را در $y=\pi/4$ رسم کنید.
2. **مشتق دوم این تابع نسبت به $x$ و $y$** را رسم کنید.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0, 2*np.pi,20)
y = np.sin(x)

yp = np.diff(y)/np.diff(x)
plt.plot(x, y)
plt.plot(x[:-1], yp)
```
